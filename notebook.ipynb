```python
#Day 1 Part 1
f = open('1.txt')
lines = f.readlines()
lines = [line.strip() for line in lines]
digit_lines = [[a for a in line if a.isnumeric()] for line in lines]
code_lines = [int(line[0]) * 10 + int(line[len(line) - 1]) for line in digit_lines]
print(sum(code_lines))
```

```python
#Day 1 Part 2
import math
places = { 'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9 }
"""lines = ["two1nine",
"eightwothree",
"abcone2threexyz",
"xtwone3four",
"4nineeightseven2",
"zoneight234",
"7pqrstsixteen"]"""
#lines = ["bxlbtwotwoeight9twoseveneight"]
place_lines = []
for line in lines:
    first = min(places.keys(), key=lambda x: line.find(x) if line.find(x) > -1 else math.inf)
    if line.find(first) != -1:
        line = line.replace(first, str(places[first]) + first, 1)
    #We reverse the string to find the last occurence of a number
    line = line[::-1]
    last = min([k[::-1] for k in places.keys()], key=lambda x: line.find(x) if line.find(x) > -1 else math.inf)
    if line.find(last) != -1:
        line = line.replace(last, last + str(places[last[::-1]]), 1)
    place_lines.append(line[::-1])
digit_lines = [[a for a in line if a.isnumeric()] for line in place_lines]
code_lines = [int(line[0]) * 10 + int(line[len(line) - 1]) for line in digit_lines]
print(sum(code_lines))
```

```python
f = open('2.txt')
lines = f.readlines()
lines = [line.strip() for line in lines]
ids = 0
limits = {'blue': 14, 'red': 12, 'green': 13}
current_id = 1
for game in lines:
    wrong = False
    data = game.split(":")[1]
    splits = data.split(";")
    for split in splits:
        cubes = split.split(",")
        for cube in cubes:
            _, number, color = cube.split(" ")
            if int(number) > limits[color]:
                wrong = True
                break
        if wrong:
            break
    if not wrong:
        ids += current_id
    current_id += 1
print(ids)
```

```python
powers = 0
for game in lines:
    data = game.split(":")[1]
    splits = data.split(";")
    limits = {'blue': 0, 'red': 0, 'green': 0}
    for split in splits:
        cubes = split.split(",")
        for cube in cubes:
            _, number, color = cube.split(" ")
            if int(number) > limits[color]:
                limits[color] = int(number)
    powers += limits['blue'] * limits['red'] * limits['green']
print(powers)
```

```python
def bts(tab):
    return "\n".join(["".join(c) for c in tab])

def valid_symbol(char):
    return not (char.isnumeric() or char == ".")

def keep_char(array, i, j):
    indexes = [(i, j-1), (i, j+1), (i-1, j-1), (i+1, j-1), (i-1, j+1), (i+1, j+1), (i-1, j), (i+1, j)]
    return any([valid_symbol(array[x][y]) for x, y in indexes])
    

def read_engine(txt):
    #We first add things to avoid array bounds
    array = [list("." + line.strip() + ".") for line in txt.split("\n")]
    array.insert(0, list("." * (len(array[0]))))
    array.insert(len(array), list("." * (len(array[0]))))
    #print(bts(array))
    sum = 0
    for i in range(len(array)):
        keep = False
        number = ""
        for j in range(len(array[0])):
            charat = array[i][j]
            if charat.isnumeric():
                number += charat
                if keep_char(array, i, j):
                    keep = True
            elif number != "":
                if keep:
                    sum += int(number)
                    #print("Adding", number, "total", sum)
                number = ""
                keep = False
    return sum

test = """467..114..
...*......
..35..633.
......#...
617*......
.....+.58.
..592.....
......755.
...$.*....
.664.598.."""
#print(read_engine(test), "must be 4361")
data = open("3.txt").read()
print(read_engine(data))
```

```python
def valid_symbol(char):
    return (not char.isnumeric()) and char == "*"

def keep_char_out(array, i, j):
    indexes = [(i, j-1), (i, j+1), (i-1, j-1), (i+1, j-1), (i-1, j+1), (i+1, j+1), (i-1, j), (i+1, j)]
    for x, y in indexes:
        if valid_symbol(array[x][y]):
            return (x, y)
    return False
    

def gear_engine(txt):
    #We first add things to avoid array bounds
    array = [list("." + line.strip() + ".") for line in txt.split("\n")]
    array.insert(0, list("." * (len(array[0]))))
    array.insert(len(array), list("." * (len(array[0]))))
    sum = 0
    dict = {}
    for i in range(len(array)):
        keep = False
        number = ""
        for j in range(len(array[0])):
            charat = array[i][j]
            if charat.isnumeric():
                number += charat
                kco = keep_char_out(array, i, j)
                if kco:
                    keep = kco
            elif number != "":
                if keep:
                    if not keep in dict:
                        dict[keep] = []
                    dict[keep].append(int(number))
                number = ""
                keep = False
    for items in dict.values():
        if len(items) == 2:
            sum += items[0] * items[1]
    return sum
test = """467..114..
...*......
..35..633.
......#...
617*......
.....+.58.
..592.....
......755.
...$.*....
.664.598.."""
print(gear_engine(test), "must be 467835")
data = open("3.txt").read()
print(gear_engine(data))
```

```python
def stratchcards(txt):
    lines = txt.split("\n")
    splitted = [line.strip().split(":")[1].split("|") for line in lines]
    total = 0
    for win, number in splitted:
        n_win = [int(n) for n in win.split(" ") if n.isnumeric()]
        n_number = [int(n) for n in number.split(" ") if n.isnumeric()]
        wins = len([n for n in n_number if n in n_win])
        if wins > 0:
            total += 2 ** (wins - 1)
    return total

test = """Card 1: 41 48 83 86 17 | 83 86  6 31 17  9 48 53
Card 2: 13 32 20 16 61 | 61 30 68 82 17 32 24 19
Card 3:  1 21 53 59 44 | 69 82 63 72 16 21 14  1
Card 4: 41 92 73 84 69 | 59 84 76 51 58  5 54 83
Card 5: 87 83 26 28 32 | 88 30 70 12 93 22 82 36
Card 6: 31 18 13 56 72 | 74 77 10 23 35 67 36 11"""

print(stratchcards(test), "must be 13")
print(stratchcards(open("4.txt").read().strip()))
```

```python
def copies_stratchcards(txt):
    lines = txt.split("\n")
    splitted = [line.strip().split(":")[1].split("|") for line in lines]
    dict = {a:1 for a in range(len(splitted))}
    i = 0
    for win, number in splitted:
        n_win = [int(n) for n in win.split(" ") if n.isnumeric()]
        n_number = [int(n) for n in number.split(" ") if n.isnumeric()]
        wins = len([n for n in n_number if n in n_win])
        for j in range(wins):
            dict[j+i+1] += dict[i]
        i += 1
    return sum(dict.values())

test = """Card 1: 41 48 83 86 17 | 83 86  6 31 17  9 48 53
Card 2: 13 32 20 16 61 | 61 30 68 82 17 32 24 19
Card 3:  1 21 53 59 44 | 69 82 63 72 16 21 14  1
Card 4: 41 92 73 84 69 | 59 84 76 51 58  5 54 83
Card 5: 87 83 26 28 32 | 88 30 70 12 93 22 82 36
Card 6: 31 18 13 56 72 | 74 77 10 23 35 67 36 11"""

print(copies_stratchcards(test), "must be 30")
print(copies_stratchcards(open("4.txt").read().strip()))
```

```python
def ranged_fertilizer(text):
    def fill_stage(ranges, actions):
        print("Filling stage w ranges", ranges, "w action", actions)
        acted_ranges = []
        to_act_ranges = ranges
        for (dest, src, size) in actions:
            post_to_act_ranges = []
            eoa = src + size
            for (start, stop) in to_act_ranges:
                if start >= eoa or stop <= src:#There is no interlap between the two ranges
                    post_to_act_ranges.append((start, stop))
                else:
                    act_start, act_end = dest + start - src, stop
                    if src >= start:#pre segment
                        post_to_act_ranges.append((start, src))
                        act_start = dest
                    if eoa < stop:#post segment
                        post_to_act_ranges.append((eoa, stop))
                        act_end = eoa
                    acted_ranges.append((act_start, act_end + dest - src))
            to_act_ranges = post_to_act_ranges
            print("Post action", acted_ranges, to_act_ranges)
        return acted_ranges + to_act_ranges
    def range(list):
        return [(a,b) for a, b in zip(list[::2], list[1::2])]
    zones = text.strip().split("\n\n")
    seeds = range([int(n) for n in zones[0].split(": ")[1].split(" ")])[:2]
    ranges = [(a, a+b) for (a,b) in seeds]
    for zone in zones[1:]:
        lines = [[int(n) for n in line.split(" ")] for line in zone.split(":\n")[1].split("\n")]
        ranges = fill_stage(ranges, lines)
    #return [(a, b) for (a, b) in ranges if a == 0]
    print(ranges)
    return min([a for (a,b) in ranges])

#print(ranged_fertilizer(test))
print(ranged_fertilizer(open("5.txt").read()))
```
